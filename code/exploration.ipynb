```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rnn = open("rnn.txt", "r")
rnn_lines = rnn.readlines()
```

```python
rnn_metrics_dict = {}
for line in rnn_lines:
    if "Lookback" in line:
        lookback = int(line.replace("Lookback = ", "").replace("\n", ""))
        rnn_metrics_dict[lookback] = {}
    elif "Average" in line:
        average = float(line.replace("Average: ", "").replace("\n", ""))
        if "Average" not in rnn_metrics_dict[lookback]:
            rnn_metrics_dict[lookback]["Average"] = []
        rnn_metrics_dict[lookback]["Average"].append(average)
    elif "Magnitude" in line:
        magnitude = float(line.replace("Magnitude: ", "").replace("\n", ""))
        if "Magnitude" not in rnn_metrics_dict[lookback]:
            rnn_metrics_dict[lookback]["Magnitude"] = []
        rnn_metrics_dict[lookback]["Magnitude"].append(magnitude)
```

```python
rnn_lookback_2_avg = rnn_metrics_dict[2]["Average"]
rnn_lookback_2_mag = rnn_metrics_dict[2]["Magnitude"]

rnn_lookback_10_avg = rnn_metrics_dict[10]["Average"]
rnn_lookback_10_mag = rnn_metrics_dict[10]["Magnitude"]
```

```python
gru_metrics_dict = {}
gru = open("gru.txt", "r")
gru_lines = gru.readlines()
for line in gru_lines:
    if "Lookback" in line:
        lookback = int(line.replace("Lookback = ", "").replace("\n", ""))
        gru_metrics_dict[lookback] = {}
    elif "Average" in line:
        average = float(line.replace("Average: ", "").replace("\n", ""))
        if "Average" not in gru_metrics_dict[lookback]:
            gru_metrics_dict[lookback]["Average"] = []
        gru_metrics_dict[lookback]["Average"].append(average)
    elif "Magnitude" in line:
        magnitude = float(line.replace("Magnitude: ", "").replace("\n", ""))
        if "Magnitude" not in gru_metrics_dict[lookback]:
            gru_metrics_dict[lookback]["Magnitude"] = []
        gru_metrics_dict[lookback]["Magnitude"].append(magnitude)
```

```python
gru_lookback_2_avg = gru_metrics_dict[2]["Average"]
gru_lookback_2_mag = gru_metrics_dict[2]["Magnitude"]

gru_lookback_10_avg = gru_metrics_dict[10]["Average"]
gru_lookback_10_mag = gru_metrics_dict[10]["Magnitude"]
```

```python
x_rnn = rnn_lookback_2_mag[:896038 // 10]
x_gru = gru_lookback_2_mag[:896038 // 10]
```

```python
plt.plot(x_rnn, label="RNN")
plt.plot(x_gru, label="GRU")
plt.legend()
plt.title("Magnitude of RNN vs GRU")
plt.ylabel("Magnitude")
plt.xlabel("Time")
```
